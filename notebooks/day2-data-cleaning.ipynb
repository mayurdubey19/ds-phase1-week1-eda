```python
import pandas as pd

import numpy as np



df = pd.read_excel(

    "/kaggle/input/sample-superstore/Sample - Superstore.xls",

    engine="xlrd"

)

```

```python
df.head()
df.shape
df.dtypes
```

**Standardize Column Names**

```python
df.columns = (

    df.columns

      .str.strip()

      .str.lower()

      .str.replace(" ", "_")

)

```

```python
df.columns
```

**Find Missing Values**

```python
df.isnull().sum().sort_values(ascending=False)

```

```python
(df.isnull().mean() * 100).round(2)
```

**Handle Missing Values**

```python
# Drop rows with missing critical fields

df = df.dropna(subset=["sales", "profit"])



# Fill categorical nulls

df["postal_code"] = df["postal_code"].fillna("UNKNOWN")

```

****Normalize data types****

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df["sales"] = pd.to_numeric(df["sales"], errors="coerce")

df["profit"] = pd.to_numeric(df["profit"], errors="coerce")

```

```python
df.dtypes
```

**Find count of duplicates**

```python
df.duplicated().sum()
```

**Removed duplicates**

```python
df = df.drop_duplicates()
```

```python
df.shape
```

**Identify Outliers**

```python
df[["sales", "profit"]].describe()

```

**Clean Outliers**

```python
Q1 = df["sales"].quantile(0.25)
print(Q1)
Q3 = df["sales"].quantile(0.75)
print(Q3)
IQR = Q3 - Q1
df = df[(df["sales"] >= Q1 - 1.5 * IQR) &

        (df["sales"] <= Q3 + 1.5 * IQR)]
df.head(2)
df.shape
```

**Save Cleaned data**

```python
df.to_csv("/kaggle/working/cleaned_superstore.csv", index=False)
```
